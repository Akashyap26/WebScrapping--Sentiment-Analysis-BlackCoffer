Pre-Processing the Text data for Sentimenanalysis

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import SnowballStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import stopwords
import re
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer() 

def preprocess(sentence):
    sentence=str(sentence)
    sentence = sentence.lower()
    sentence=sentence.replace('{html}',"") 
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url=re.sub(r'http\S+', '',cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_num)  
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopWordList]
    stem_words=[stemmer.stem(w) for w in filtered_words]
    lemma_words=[lemmatizer.lemmatize(w) for w in stem_words]
    return " ".join(filtered_words)


dataM['cleanText']=dataM['Paragraph'].map(lambda s:preprocess(s))
dataM
```

Writing Function Block for Each of the features 

```python
from operator import sub
Positive_words = [line.strip() for line in open("/content/drive/MyDrive/BlackCoffer/MasterDictionary/positive-words.txt", 'r')] 
Negetive_Words = [line.strip() for line in open("/content/drive/MyDrive/BlackCoffer/MasterDictionary/negative-words.txt", 'r',encoding="ISO-8859-1")]

def positivity_score(text):
    words = text.lower().split()
    num_positive_words = sum([1 for word in words if word in Positive_words])
    positivity_score = num_positive_words / len(words)
    return positivity_score

def negativity_score(text):
    neg_count = 0
    word_list = text.lower().split()

    for word in word_list:
        if word in Negetive_Words:
            neg_count += 1

    return neg_count/len(word_list)


def polarity_score(positivity_score,negativity_score):
  plscore = ((positivity_score - negativity_score) / (positivity_score + negativity_score)+ 0.000001)
  return plscore

def Subjectivity_Score(text,pscore,nscore):
  word_count = len(nltk.word_tokenize(text))
  subscore = (pscore + nscore)/((word_count) + 0.000001)
  return subscore


def avg_sentence_length(text):
    sentences = nltk.sent_tokenize(text)
    total_words = 0
    for sentence in sentences:
        total_words += len(nltk.word_tokenize(sentence))
    return (total_words / len(sentences))

```

```python
#Tokanizer
def tokenizer(text):
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(text)
    filtered_words = list(filter(lambda token: token not in stopWordList, tokens))
    return filtered_words
    
```

```python
def complex_word_percentage(text):
    # Tokenize the text into words
    words = nltk.word_tokenize(text)
    
    # Define a list of vowels
    vowels = ['a', 'e', 'i', 'o', 'u']
    
    # Initialize a count for complex words
    complex_word_count = 0
    
    # Loop through each word in the text
    for word in words:
        # Count the syllables in the word
        syllable_count = 0
        for letter in word:
            if letter.lower() in vowels:
                syllable_count += 1
        
        # Determine if the word is complex based on the number of syllables
        if syllable_count > 2:
            complex_word_count += 1
    
    # Calculate the percentage of complex words
    complex_word_percentage = (complex_word_count / len(words)) * 100
    
    return complex_word_percentage


```

```python
def fog_index(avg_sentence_length,complex_word_percentage):
  fx = 0.4 * (avg_sentence_length + complex_word_percentage)
  return fx

def Average_number_of_words_per_sentence(text):
  sentences = nltk.sent_tokenize(text)
  word_counts = [len(nltk.word_tokenize(sentence)) for sentence in sentences]
  avg_words = (sum(word_counts) / len(word_counts))
  return avg_words

def complex_word_count(text):
    tokens = tokenizer(text)
    complexWord = 0
    
    for word in tokens:
        vowels=0
        if word.endswith(('es','ed')):
            pass
        else:
            for w in word:
                if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u'):
                    vowels += 1
            if(vowels > 2):
                complexWord += 1
    return complexWord
```

```python
def word_count(text):
  tokens = tokenizer(text)
  return len(tokens)

def syllable(text):
  syllable_count = 0
  word_count = len(nltk.word_tokenize(text))
  syllables = nltk.corpus.cmudict.dict()
  words = nltk.word_tokenize(text)
  for word in words:
      if word.lower() in syllables:
          syllable_count += max([len(list(y for y in x if y[-1].isdigit())) for x in syllables[word.lower()]])
  syllables_per_word = (syllable_count / word_count)
```

```python
import nltk
from nltk.corpus import cmudict

# load the CMU Pronouncing Dictionary
nltk.download('cmudict')
pronouncing_dict = cmudict.dict()

def count_syllables(word):
    """
    Count the number of syllables in a word using the CMU Pronouncing Dictionary in NLTK.
    """
    # if word not found in pronouncing dictionary, return 0
    if word.lower() not in pronouncing_dict:
        return 0
    
    # count the number of syllables in the word
    syllable_count = [len(list(y for y in x if y[-1].isdigit())) for x in pronouncing_dict[word.lower()]][0]
    return syllable_count
```

```python
def count_personal_pronouns(text):
    personal_pronouns = ['i', 'me', 'my', 'mine', 'you', 'your', 'yours', 
                         'he', 'him', 'his', 'she', 'her', 'hers', 
                         'it', 'its', 'we', 'us', 'our', 'ours', 
                         'they', 'them', 'their', 'theirs']
    
    tokens = nltk.word_tokenize(text.lower())
    personal_pronoun_count = sum(1 for token in tokens if token in personal_pronouns)
    
    return personal_pronoun_count
```

```python
def avg_word_length(text):
    words = nltk.word_tokenize(text)
    word_lengths = [len(word) for word in words]
    avg_length = sum(word_lengths) / len(word_lengths)
    return avg_length
```

Applying the Function Block for each of the pandas Data Frame

```python
dataM['POSITIVE_SCORE'] = dataM['cleanText'].apply(positivity_score)
dataM['NEGATIVE_SCORE'] = dataM['cleanText'].apply(negativity_score)
dataM['POLARITY_SCORE'] = np.vectorize(polarity_score)(dataM['POSITIVE_SCORE'],dataM['NEGATIVE_SCORE'])
dataM['SUBJECTIVITY_SCORE'] = np.vectorize(Subjectivity_Score)(dataM['cleanText'],dataM['POSITIVE_SCORE'],dataM['NEGATIVE_SCORE'])
dataM['AVG_SENTENCE_LENGTH'] = np.vectorize(avg_sentence_length)(dataM['Paragraph'])
dataM['PERCENTAGE_OF_COMPLEX_WORDS'] =dataM['cleanText'].apply(complex_word_percentage)
dataM['FOG_INDEX'] = np.vectorize(fog_index)(dataM['AVG_SENTENCE_LENGTH'],dataM['PERCENTAGE_OF_COMPLEX_WORDS'])
dataM['AVG_NUMBER_OF_WORDS_PER_SENTENCE'] = dataM['Paragraph'].apply(Average_number_of_words_per_sentence)
dataM['COMPLEX_WORD_COUNT'] = dataM['cleanText'].apply(complex_word_count)
dataM['WORD_COUNT'] = dataM['cleanText'].apply(word_count)
dataM['SYLLABLE_PER_WORD'] = dataM['cleanText'].apply(count_syllables)
dataM['PERSONAL_PRONOUNS'] = dataM['cleanText'].apply(count_personal_pronouns)
dataM['AVG_WORD_LENGTH'] = dataM['cleanText'].apply(avg_word_length)
```
